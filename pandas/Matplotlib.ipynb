### Title and Lables

```python
fig3,ax3 = plt.subplots(1,1)
```

```python
# the functional approach

plt.plot([i for i in range(10)],color='black', linestyle='--',label='a')
plt.plot([i**2 for i in range(10)],color='cyan', linestyle='--',label='b')
plt.legend()
plt.title('lines')
plt.xlabel('x_label')
plt.ylabel('y_label')
```

```python
## object oriented way
```

```python
fig5,ax5 = plt.subplots()
ax5.plot([i for i in range(10)],color='g',ls='--',label='a')
ax5.plot([i**2 for i in range(10)],color='magenta',ls='-',label='b')
ax5.set_title('my graph title')
ax5.set_xlabel('x_label')
ax5.set_ylabel('y_label')
ax5.legend()
```

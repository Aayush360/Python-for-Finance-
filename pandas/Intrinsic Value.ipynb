### Revenue

```python
df_rev=pd.read_csv('datas/50-rev.csv',index_col=0)
```

```python
df_rev
```

```python
rev_df_ss=df_rev.loc[2009:2019]
```

```python
rev_df_ss=rev_df_ss.transpose()
```

```python
# calculating 10 years growth rate
```

```python
rev_df_ss['10Y CAGR']=(rev_df_ss[2019]/rev_df_ss[2009])**(1/10 )-1
```

```python
rev_df_ss
```

```python
rev_df_ss['5Y CAGR']=(rev_df_ss[2019]/rev_df_ss[2014])**(1/5)-1
```

```python
rev_df_ss['1Y CAGR']=(rev_df_ss[2019]/rev_df_ss[2018])-1
```

```python
rev_df_ss.head()
```

```python
data=rev_df_ss[['10Y CAGR','5Y CAGR','1Y CAGR']]
```

```python
data.sort_values('1Y CAGR',ascending=False)
```

```python
data.sort_values('10Y CAGR',ascending=False)
```

```python
data.to_csv('result-revenue.csv')
```

### Earnings Per Share

```python
# is the bottom line of the income statement
```

```python
# we are interested in the comapany that have growth in the revenue but that growth in the revenue should turn into
# growth in the earnings because if earnings does not grow as the revenue grows then they are spending too much to
# get that growth

# should ensure that growth in the revenue should give growth in the earnings
# we only care that the revenue turn into earnings i,e we only care about the growth rate
```

```python
eps = pd.read_csv('datas/50-eps.csv',index_col=0)
eps.head()
```

```python
eps.dtypes # checking the datatype,, 
```

```python
eps = eps.transpose()
```

```python
eps.head()
```

```python
eps['10Y CAGR'] = (eps[2019]/eps[2009])**(1/10)-1
eps['5Y CAGR'] = (eps[2019]/eps[2014])**(1/5)-1
eps['1Y CAGR'] = (eps[2019]/eps[2018])**(1/1)-1
```

```python
eps.head()
```

```python
eps.sort_values('10Y CAGR',ascending=False)
```

```python
eps.sort_values('5Y CAGR',ascending=False)
```

```python
eps.sort_values('1Y CAGR',ascending=False)
```

```python
data = eps[['10Y CAGR','5Y CAGR','1Y CAGR']]
```

```python
data.to_csv('result-eps.csv')
```

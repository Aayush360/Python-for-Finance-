```python
gdp_df=pd.read_csv('gdp.csv',index_col='Country (or territory)')
```

```python
gdp_df.head()
```

```python
col1=gdp_df.columns

col2=gdp_df.iloc[0]
```

```python
col2=col2.values.tolist()
col2
```

```python
col1=col1.tolist()
col1
```

```python
l1=[]
for i in range(len(col1)):
    if col1[i]!=col2[i]:
        l1.append(col1[i]+' '+col2[i])
    else:
        l1.append(col1[i])
    
```

```python
gdp_df.columns=l1
gdp_df = gdp_df[1:]
```

```python
gdp_df.head()
```

```python
## creating a view and modify it actually will reflect in the original data
```

```python
gdp_df_view_china = gdp_df.loc['China *']
```

```python
gdp_df_view_china
```

```python
gdp_df_view_china['CIA[6].1 Year']='2020'
```

```python
# the change is reflected in the original dataframe i.e gdp_df

gdp_df
```

```python
# transposing a dataframe

gdp_df.T
```

```python
# difference between the view and the copy

# creating a view and making a changes in the view will actually reflect in the original data
# but creating a copy will only make a change in the copied data instead of the original data
```

```python
gdp_00 = pd.read_csv('datas/gdp-00s.csv',index_col=0)
gdp_00.head()
```

```python
gdp_10 = pd.read_csv('datas/gdp-10s.csv',index_col=0)
gdp_10.head()
```

```python
gdp = pd.concat([gdp_00,gdp_10],axis=1)
gdp.head()
```

```python
# to drop all rows with nan

gdp.dropna(axis=0,inplace=True)
gdp.dropna()
```

```python
# calculate gdp growth from 2000-2019 in percentage in column named 20y-pct

gdp['20y-pct']=(gdp['2019']-gdp['2000'])*100/gdp['2000']
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'a':[1,2,3,4],'b':[5,6,7,8]})
df2 = pd.DataFrame({'b':[5,6,7,8],'c':[9,10,11,12]})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2]) # it is like doing a union in sql

# default concat is by axis=0 i.e by the row
```

```python
pd.concat([df1,df2],axis=1) # concat by column
```

```python
pd.merge(df1,df2,on='b') # inner join
```

```python
pd.merge(df1,df2,how='left',on='b')
```

```python
df3 = pd.DataFrame({'b':[8,9,10,11],'c':[1,2,3,4]})
```

```python
pd.merge(df2,df3,on='b') # since there is only 1 key that is a match
```

```python
gdp_df=pd.read_csv('gdp.csv',index_col='Country (or territory)')
```

```python
gdp_df.head()
```

```python
col1=gdp_df.columns

col2=gdp_df.iloc[0]
```

```python
col2=col2.values.tolist()
col2
```

```python
col1=col1.tolist()
col1
```

```python
l1=[]
for i in range(len(col1)):
    if col1[i]!=col2[i]:
        l1.append(col1[i]+' '+col2[i])
    else:
        l1.append(col1[i])
    
```

```python
gdp_df.columns=l1
gdp_df = gdp_df[1:]
```

```python
gdp_df.head()
```

```python
## creating a view and modify it actually will reflect in the original data
```

```python
gdp_df_view_china = gdp_df.loc['China *']
```

```python
gdp_df_view_china
```

```python
gdp_df_view_china['CIA[6].1 Year']='2020'
```

```python
# the change is reflected in the original dataframe i.e gdp_df

gdp_df
```

```python
# transposing a dataframe

gdp_df.T
```

```python
# difference between the view and the copy

# creating a view and making a changes in the view will actually reflect in the original data
# but creating a copy will only make a change in the copied data instead of the original data
```

```python
 gdp_df_copy_china = gdp_df.loc['China *'].copy()
```

```python
gdp_df_copy_china
```

```python
gdp_df_copy_china['CIA[6].1 Year']='2020'
```
